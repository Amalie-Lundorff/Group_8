## Figure and Axes

Why do we have two separate plots? The reason is that, by default, the `plot` method creates a new \emph{figure} instead of putting them inside a single figure. In order to avoid it, we need to either create an [Axes](https://matplotlib.org/api/axes_api.html) and tell `plot` to use that axes. What is an `axes`? See this illustration. 

<img src="https://matplotlib.org/1.5.1/_images/fig_map.png" alt="figure, axes, and axis" style="width: 500px;"/>


A figure can contain multiple axes ([link](https://matplotlib.org/gallery/subplots_axes_and_figures/subplot_demo.html#sphx-glr-gallery-subplots-axes-and-figures-subplot-demo-py)). The figure below contains two axes:

![multipleaxes](https://matplotlib.org/_images/sphx_glr_subplot_demo_001.png)

and an axes can contain multiple plots ([link](https://matplotlib.org/gallery/api/two_scales.html#sphx-glr-gallery-api-two-scales-py)). 

![multipleplots](https://matplotlib.org/_images/sphx_glr_two_scales_001.png)

Conveniently, when you call `plot` method, it creates an axes and returns it to you

```python
ax = death_df.plot(x='X', y='Y', s=2, c='black', kind='scatter', label='Deaths')
```

```python
ax
```

This object contains all the information and objects in the plot we see. Whatever we want to do with this axes (e.g., changing x or y scale, overlaying other data, changing the color or size of symbols, etc.) can be done by accessing this object. 

Then you can pass this axes object to another plot to put both plots in the same axes. Note `ax=ax` in the second plot command. It tells the plot command _where_ to draw the points. 

```python
ax = death_df.plot(x='X', y='Y', s=2, c='black', alpha=0.5, kind='scatter', label='Deaths')
pump_df.plot(x='X', y='Y', kind='scatter', c='red', s=8, label='Pumps', ax=ax)
```

Although simply invoking the `plot()` command is quick and easy when doing an exploratory data analysis, it is usually better to be formal about figure and axes objects. 

Here is the recommended way to create a plot. Call the `subplots()` method (see https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.subplots.html) to get the figure and axes objects explicitly. 

As you can see below, `subplots()` creates an empty figure and returns the figure and axes object to you. Then you can fill this empty canvas with your plots. Whatever manipulation you want to make about your figure (e.g., changing the size of the figure) or axes (e.g., drawing a new plot on it) can be done with `fig` and `ax` objects. So whenever possible, use this method! 

**Now, can you use this method to produce the same plot just above?**

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

# your code here
#Plot the deaths dataset on the Axes
    #I am going to use the plot method to create a scatter plot for the deaths dataset (death_df:)

#Plot deaths data
ax=death_df.plot(
    x='X', 
    y='Y', 
    kind='scatter', 
    s=2, 
    c='black', 
    alpha=0.5, 
    label='Deaths', 
    ax=ax
)
#overlay the pumps dataset on the same Axes
pump_df.plot(
    x='X', 
    y='Y', 
    kind='scatter', 
    c='pink', 
    s=8, 
    label='Pumps', 
    ax=ax
)
#show the combined plot
plt.show()
```

## Voronoi diagram

Let's try the Voronoi diagram. You can use the `scipy.spatial.Voronoi` and `scipy.spatial.voronoi_plot_2d` from `scipy`, the *scientific python* library. 

```python
from scipy.spatial import Voronoi, voronoi_plot_2d
```

Take a look at the documentation of [Voronoi](https://docs.scipy.org/doc/scipy-0.18.1/reference/generated/scipy.spatial.Voronoi.html#scipy.spatial.Voronoi) and [voronoi_plot_2d](https://docs.scipy.org/doc/scipy-0.18.1/reference/generated/scipy.spatial.voronoi_plot_2d.html) and 

**Q3: produce a Voronoi diagram that shows the deaths, pumps, and voronoi cells**

```python
# you'll need this
points = pump_df.values
points
```

```python
# TODO: your code here
#1: import the necessary libraries
from scipy.spatial import Voronoi, voronoi_plot_2d

#2: Extract the pump coordinates from the pump_df
points = pump_df.values
    #we understand from the given information that the pump coordinates are stored in pump_df
    #the .values extracts the NumPy array of coordinate: 
        #this gives us the raw coordinates of the pumps. these are the points used to calculate the Voronoi diagram. 

#3: Computing the Voronoi diagram
vor=Voronoi(points)
    #we pass the pump coordinates (points) to the voronoi function to compute the voronoi cells: 
        #this creates a Voronoi object, which contains all the information about the cells and regions. 

#4: Plotting the Voronoi diagram
voronoi_plot_2d(vor)
    #we use the voronoi_plot_2d function to visualize the Voronoi diagram: 
        #this function takes the vor object as an argument and plots the diagram.

#5: overlay the pump and death data
#plotting pumps in pink
plt.scatter(pump_df['X'], pump_df['Y'], color='pink', label='Pumps',s=80)

#plotting deaths in purple
plt.scatter(death_df['X'], death_df['Y'], color='purple', label='Deaths',s=20, alpha=0.5)
    #we overlay the pump and death data on the Voronoi diagram for better context, we use matplotlib's scatter to plot the points.
    #we use alpha to make the death points semi-transparent, so we can see the Voronoi diagram underneath.
#add a legend
plt.legend()
    #we add a legend to the plot to help distinguish between the pumps and deaths.
plt.show()



```

Conclusion: 
This plot visualizes the relationship between cholera deaths and and water pumps using Voronoi diagram. The black lines represent Vornoi cells, which divide the space based on proximity to the pumps, meaning each cell contains the area closest to a specific pump. the orange points mark the locations of the pumps, while the purple points represent individual cholera deaths. By overlaying the death locations into the Voronai, we can observe that deaths are not evenly distributed but are instead highly concentrated around certian pumps. this suggests a strong correlation between pump locations and the spread of cholera. 

The clustering of deaths within particular Voronoi regions indicates that residents relying on specific pumps were at greater risk, reinforcing the idea that contaminated water sources were the primary cause of the outbreak.

#1. voronoi regions: 
    the black lines represents the voronoi regions. Each region shows the area closer to one pump than any other pump. 

    These regions divide the space based on the proximity to the pumps, which is what the voronoi diagram is meant to do. 

#2. pump locations (orange): 
    the orange points (pumps) are correctly located at the voronoi region centers, as these are the generators of the diagram. 

#3. Death locations (purple): 
    The purple points represent the locations of deaths. they are plotted on the top of the voronoi regions to show thier spatial distribution relative to the pumps. 

#4. legend: 
    the legend in the top right corner clearly labels the symbold (pumps and deaths)
